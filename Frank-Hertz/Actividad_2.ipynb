```python
V160 = data7[1]
I160 = data7[2]
V160 = np.array(V160, dtype=float)
I160 = np.array(I160, dtype=float)
V160 = V160[~np.isnan(V160)]
I160 = I160[~np.isnan(I160)]
V160
```

```python
V180 = data7[15]
I180 = data7[16]
V180 = np.array(V180, dtype=float)
I180 = np.array(I180, dtype=float)
V180 = V180[~np.isnan(V180)]
I180 = I180[~np.isnan(I180)]
V180
```

```python
min_160 = []
min_170 = []
min_180 = []

for i in range(len(V160)):
    if i%2 == 1:
        min_160.append(V160[i])
for i in range(len(V170)):
    if i%2 == 1:
        min_170.append(V170[i])
for i in range(len(V180)):
    if i%2 == 1:
        min_180.append(V180[i])

diff_160 = []
n_160 = []
for i in range(1,len(min_160)):
        diff_160.append(min_160[i] - min_160[i-1])
        n_160.append(i)
diff_170 = []
n_170 = []
for i in range(1,len(min_170)):
        diff_170.append(min_170[i] - min_170[i-1])
        n_170.append(i)
diff_180 = []
n_180 = []
for i in range(1,len(min_180)):
        diff_180.append(min_180[i] - min_180[i-1])
        n_180.append(i)
len(n_160),len(n_170),len(n_180)
```

```python
diff = [diff_160,diff_170,diff_180]
n = [n_160,n_170,n_180]
col = ['blue','orange','r']
T = ['T = 160°C','T = 170°C','T = 180°C']
c = []
c_err = []
E_a = []
```

```python
plt.rc('font', family='Times New Roman') 
fig, axs = plt.subplots(2, 1, figsize=(10, 7), gridspec_kw={'height_ratios': [1.5, 1]},sharex=True)
fig.subplots_adjust(hspace=0)
fig.suptitle("Actividad 2.7", fontsize=18)
for i in range(len(n)):
    coef,cov = np.polyfit(n[i],diff[i],1,cov=True)
    coef_err = np.sqrt(np.diagonal(cov))
    model = np.poly1d(coef)
    c.append(coef)
    c_err.append(coef_err)
    E_a.append(model(0.5))
    axs[0].plot(n[i],model(n[i]),label=T[i],color=col[i])
    axs[1].scatter(n[i],diff[i]-model(n[i]),color=col[i])
axs[0].scatter(n_160,diff_160,color='blue')
axs[0].scatter(n_170,diff_170,marker='x',color='orange')
axs[0].scatter(n_180,diff_180,marker='+',color='r')
axs[0].set_ylabel('Distancia entre mínimos (eV)', fontsize=16)
axs[0].legend()
axs[0].grid(True)
axs[1].set_xlabel('Orden del mínimo',fontsize=16)
axs[1].axhline(y=0, color='black', linestyle='--')
axs[1].set_ylabel('Residual')
axs[1].grid(True)
```
