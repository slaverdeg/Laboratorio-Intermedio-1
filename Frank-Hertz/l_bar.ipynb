```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 299798458 #m/s
kb = 1.380649*1e-23 #J/K
h = 6.62607015*1e-34 #J*s
r = 1.5*1e-10 #m
sigma = np.pi*(r**2)
l = lambda T : (kb*T) / (np.sqrt(2) * 8.7*(10**(9 - (3110/T))) * sigma)
T = np.linspace(300,500)

plt.rc('font', family='Times New Roman') 
plt.figure(figsize = (10, 7))
plt.plot(T,l(T),color='r')
plt.title("Camino medio libre", fontsize=18)
plt.ylabel(r'$\bar{l}$ (m)', fontsize=16)
plt.xlabel('T (K)',fontsize=16)
plt.grid(True)
plt.legend()
```
