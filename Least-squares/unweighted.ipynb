```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Unweighted fit.

```python
def unweighted_fit(x,y):
    
    sum_x2 = np.sum(x**2)
    sum_y = np.sum(y)
    sum_x = np.sum(x)
    sum_xy = np.sum(x*y)
    N = len(x)
    Delta = (N * sum_x2) - sum_x**2
    
    m = (N * sum_xy - sum_x * sum_y) / Delta
    c = (sum_x2 * sum_y - sum_x * sum_xy) / Delta
    
    CU = np.sqrt( (1/(N-2)) * np.sum((y - m*x - c)**2) )
    m_err = CU * np.sqrt(N / Delta)
    c_err = CU * np.sqrt(sum_x2 / Delta)
    
    return (m,c) , (m_err,c_err) ,CU
```

# 5.3. 

```python
V_err = 0.01
V = np.array([0.98,1.98,2.98,3.97,4.95,5.95,6.93,7.93,8.91])
I = np.array([10.,20.,30.,40.,50.,60.,70.,80.,90.])

coef, error, CU = unweighted_fit(I,V)
coef, error ,CU
```

m = $(9908 \pm 9) \times 10^{-5}$ $k\Omega$ \
c = $(0 \pm 5) \times 10^{-3}$ mV

$\alpha_{CU} = 0.007$ \
Comparado con el error experimental de 0.01, podemos evidenciar que es menor y por tanto el ajuste es apropiado.

```python
x = I 
y = V
m = coef[0]
c = coef[1]
y_ = m*x + c
incer_y = V_err

fig, axs = plt.subplots(2, 1, figsize=(10, 7), gridspec_kw={'height_ratios': [1.5, 1]},sharex=True)
fig.subplots_adjust(hspace=0)
fig.suptitle("5.3.", fontsize=18)

axs[0].scatter(x,y,marker='x',color='r',label='Data')
#axs[0].errorbar(x,y,yerr= incer_y, fmt='none', ecolor='black')
axs[0].plot(x,y_,label='Ajuste')
axs[0].set_ylabel("Voltaje (mV)", fontsize=16)
#axs[0].text(0.184,0.00152,'m = (109e-04 $\pm$ 2e-04)\nb = (0.6e-05 $\pm$ 2.8e-05)\n$R^{2} = 0.973$',fontsize=10, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'))
axs[0].legend()

axs[1].scatter(x,(y-y_)/incer_y)
axs[1].axhline(y=0, color='black', linestyle='--')
axs[1].set_xlabel(r'Corriente ($\mu$A)',fontsize=16)
axs[1].set_ylabel('Residual normalizado')
```
